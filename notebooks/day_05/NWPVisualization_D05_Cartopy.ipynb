```python
## Ploting Satelitte Image using Cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import xarray as xr

# Load the satellite dataset
dataset = xr.open_dataset('Satellite_GK2A_Fiji/gk2a_ami_le1b_ir087_fji020me_202312070830_converted.nc')
values = dataset['image_pixel_values']

# Create a new figure with a PlateCarree projection
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree(central_longitude=180))

# Add coastlines to the map
ax.coastline(linewidth=0.5)

# Add country borders
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Sets the extent using a lon/lat box
ax.set_extent([150, 200, -30, 0])

# Add gridlines with custom spacing
gl = ax.gridlines(draw_labels=True, crs=ccrs.PlateCarree(), linewidth=0.3, color='grey', linestyle='--', alpha=0.5)
gl.xlocator = plt.FixedLocator(range(-180, 181, 10))  # Longitude gridlines every 10 degrees
gl.ylocator = plt.FixedLocator(range(-90, 91, 10))    # Latitude gridlines every 10 degrees
gl.top_label = False
gl.right_labels = False
gl.xlabel_style = {'size': 50, 'color': 'blue'}
gl.ylabel_style = {'size': 5, 'color': 'blue'}

# Plot the satellite values as a heatmap
im = ax.imshow(values, extent=(values.lons.min(), values.lons.max(), values.lats.min(), values.lats.max()),
               transform=ccrs.PlateCarree(), cmap='viridis', origin='lower')

# Add colorbar
cbar = fig.color(im, ax=ax, fraction=0.03, pad=0.05)
cbar.set_label('Brightness Temperature (K)')

# Add a title
ax.set_title('GK2A Satellite Image - IR087 Channel')

plt.show()
```
