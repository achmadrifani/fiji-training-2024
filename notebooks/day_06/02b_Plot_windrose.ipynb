```python
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes
```

```python
# Read the wind data from the CSV file
df = pd.read_csv('D:/Projects/fiji-training-2024/notebooks/day_06/sample_wind_fiji_subset.csv')
```

```python
# Extract wind direction and wind speed
wind_direction = df['wind_dir']
wind_speed = df['wind_speed']
```

```python
# Create a windrose plot
ax = WindroseAxes.from_ax()
ax.bar(wind_direction, wind_speed, normed=True, bins=[2,5,10,20], opening=0.8, edgecolor='white')

# Customize the plot (optional)
ax.set_legend()
plt.title('Windrose')

# Show the plot
plt.show()
```
