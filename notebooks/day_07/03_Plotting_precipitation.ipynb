## Part 1 : Read the data
The precipitation data from an NWP model is accumulated from the first time step to the last. In order to get the amount of precipitation at a specific time, you need to subtract the precipitation at the previous time step from the current time step.

```python
import xarray as xr
```

```python
# define the file path for the first and second time step
f1 = 'D:/Projects/fiji-training-2024/data/apcp.f003'
f2 = 'D:/Projects/fiji-training-2024/data/apcp.f006'
```

```python
# read the data
ds1 = xr.open_dataset(f1, engine='cfgrib')
ds2 = xr.open_dataset(f2, engine='cfgrib')
```

```python
# check the data
ds1
```

```python
# quick plot to see the data
ds1['tp'].plot()
```

```python
# Calculate the precipitation at the time step
precip = ds2['tp'] - ds1['tp']
```

```python
# slice it to fiji region
precip_fiji = precip.sel(latitude=slice(-5,-35), longitude=slice(165,205))
precip_fiji.plot()
```
